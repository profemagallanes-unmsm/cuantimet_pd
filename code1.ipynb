```python
eduLink='https://github.com/profemagallanes-unmsm/cuantimet_pd/raw/refs/heads/main/Education%20expenditures.csv'
debtLink='https://github.com/profemagallanes-unmsm/cuantimet_pd/raw/refs/heads/main/Public%20debt.csv'

```

Use Python para leer ambas tablas y juntarlas en una sola.

```python
import pandas as pd

edu=pd.read_csv(eduLink)
debt=pd.read_csv(debtLink)

edu.info()
```

```python
debt.info()
```

```python
edu.drop(columns=['slug','ranking','date_of_information','region'],inplace=True)
debt.drop(columns=['slug','ranking','date_of_information'],inplace=True)
```

```python
edu.rename(columns={'% of GDP':'edu'},inplace=True)
debt.rename(columns={'% of GDP':'debt'},inplace=True)
```

```python
edudebt=debt.merge(edu)
edudebt
```

Cree un archivo CSV, el nuevo archivo debe tener solo las columnas pais, variables , y region

```python
edudebt.to_csv('edudebt.csv',index=False)
```
