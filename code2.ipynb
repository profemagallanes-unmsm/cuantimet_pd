```python
linkElec='https://es.wikipedia.org/wiki/Elecciones_generales_de_Per%C3%BA_de_2021'

import pandas as pd

elecs=pd.read_html(linkElec,match='Departamento')
elecs[0] # cual
```

```python
elec=elecs[0].copy() # always
```

```python
linkPob='https://es.wikipedia.org/wiki/Pobreza_en_el_Per%C3%BA'
pobs=pd.read_html(linkPob)
pobs[0]
```

```python
pob=pobs[0].copy()
```

**DEBES HABER NOTADO ERRORES**

```python
elec.info()
```

**A NUMERICO**

```python
# que columnas?
elec.columns[1::]
```

```python
elec[elec.columns[1::]]= elec.loc[:,elec.columns[1::]].apply(lambda x:x.str.replace('\s+','',regex=True)).apply(pd.to_numeric)
elec.info()
```

```python
# bye Voto extranjero
elec=elec.loc[:,['Departamento','Castillo','Electores']][:-1]
elec
```

**Aqui juntamos Lima y Callo, pues data de pobreza está así!**

```python
elec[elec.Departamento.isin(['Lima','Callao'])].sum(axis=0)
```

```python
elec.loc[25]=elec[elec.Departamento.isin(['Lima','Callao'])].sum(axis=0)
elec
```

```python
# sin esas fila
elec=elec[~elec.Departamento.isin(['Lima','Callao'])]
elec
```

```python
# renombrando
elec.loc[25,'Departamento']='Lima y Callao'
elec
```

**RESETEANDO indices**

```python
elec.reset_index(drop=True,inplace=True)
elec
```

**AHORA SI, calculando proporcion**

```python
elec['CastilloProp']=elec.Castillo/elec.Electores
elec=elec.loc[:,['Departamento','CastilloProp']] # obvio las otras
elec
```

## retomemos POBREZA

```python
pob.head()
```

```python
# solo unas columnas
pob=pob.iloc[:,[1,2,4]]
```

```python
# renombrando
pob.columns=['Departamento','pobreza','year']
pob.head()
```

```python
# usando anos !!!
pob['year']=pd.to_numeric(pob.year, errors='coerce')
pob
```

AHORA sabemos qué eliminar!

```python
pob[pob.year.isnull()]
```

```python
pob=pob[~pob.year.isnull()]
```

SACAMOS A PERu (la ultima fila)

```python
pob=pob[:-1]
pob
```

```python
# bye años!

pob.drop(columns='year',inplace=True)
```

A recuperar los numeros sin "%"

```python
pob['pobreza']=pob.pobreza.str.extract('(\d+.\d+)')
pob
```

```python
# formato ok?
pob.info()
```

```python
pob['pobreza']=pd.to_numeric(pob.pobreza)
pob.info()
```

# MERGING!!

```python
pob.merge(elec,on='Departamento',how='outer') # ver todo!!!
```

### El problema son las TILDES

```python
elec.Departamento=elec.Departamento.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('ascii')
pob.Departamento=pob.Departamento.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('ascii')
```

```python
pob.merge(elec,on='Departamento',how='outer')
```

```python
# cusco!!!
pob.Departamento=pob.Departamento.str.replace('Cuzco','Cusco')
```

```python
pob.merge(elec,on='Departamento',how='outer')
```

Para R!!

```python
elecpob=pob.merge(elec,on='Departamento',how='outer')
elecpob.to_csv('elecpob.csv',index=False)
```
